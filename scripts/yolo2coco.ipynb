```python
from pylabel import importer
```

```python
yoloclass = ['fish']

dataset = importer.ImportYoloV5(path='D:\\dis\\datasets\\fishes\\train\\labels', 
                                path_to_images='D:\\dis\\datasets\\fishes\\train\\images', 
                                cat_names=yoloclass, img_ext="jpg", name="fish")
```

```python
print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
```

```python
from IPython.display import Image, display
display(dataset.visualize.ShowBoundingBoxes(133))
display(dataset.visualize.ShowBoundingBoxes(33))
```
